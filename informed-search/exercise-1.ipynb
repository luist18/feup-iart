# Strategies for Uninformed/Informed Search 

Assuming the following search tree in which each arc displays the cost of the corresponding operator, and the nodes contain the value of the heuristic function, indicate justifying, which node is expanded next using each of the following methods:

1. Breadth-First Search
2. Depth-First Search
3. Uniform Cost Search
4. Greedy Search
5. A* Algorithm Search

### Solution

1. Node C, since B is already expanded and the BFS queue is now \[C, D, E, F, G, H\].
2. Node E, since E is the first child of B, the most expanded node.
3. Node D, since D has the lowest edge sum cost that is 3.
4. Node C, since C has the lowest heuristic function value.
5. Node G, since the function $f(G)=g(G)+h(G)$ is equal to 10 and $f(G)$ is the lowest value in the graph.
